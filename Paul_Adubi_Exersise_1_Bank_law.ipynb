```python
# I imported the numpy library for numerical computing, 
import numpy as np
# I imported the odeint from the scipy.integrate module that 
# allows us to solve ordinary differential equation numerically using the odeint function.
from scipy.integrate import odeint
# We printed the plt from the matplotlib.pyplot Python library for plotting graphs.
import matplotlib.pyplot as plt
```

```python
# We defined a function model and we passed two arguments
# y - Dependent variable - In this case, the principal amount that is changing (increasing) with respect to time.
# t - The independent variable, the time
# k = It takes a constant
def model(y,t,k):
    # Here this is what the function model returns (0.06 x 'the value of y')
    dydt = -k * y
    # We return the value of the product above
    return dydt
```

```python
# We set our initial condition for our y
y0 = 5
# We introduced an arbitrary constant = 0.3
k = 0.3
```

```python
# to plot the graph of the approximate small changes of y starting 
# from our initial conditions y0 = 5 , we set our time point from 0 to 20, splitting 0-20 into 100 equal intervals.
t = np.linspace(0,20,100)
```

```python
# we solve the differential equation using the 'odeint' function.
# the odeint() takes in the function that defines our derivative - 'model',
# it takes the intial condition y when t = 0
# and it takes our time range - 0 - 20
y = odeint(model,y0,t, args=(k,))
```

```python
# We plotted the results using matplotlib.pyplot.plt
plt.plot(t, y, label='y(t)', color='blue')
# We set our title
plt.title("Solution of dy/dt = -k * y(t) with k=0.3")
# we labeled our x axis to be time
plt.xlabel("Time (t)")
# We labeled out y axis to be our y with respect to time
plt.ylabel("y(t)")
plt.grid()
plt.legend()
# Function plots the graph
plt.show()



# KEY INSIGHTS
# We see an inverse curve observing that depreciates exponetially with respect to time from t = 0 to t = infinity
# We observe that the decay rates is observed by the constant k in this case, 0.3, larger constant leads to faster decays
# As the time progresses, the curve slopes downward, and the rate of change decrease gradually.
```
